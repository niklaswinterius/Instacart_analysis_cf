## When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Count orders per hour
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# adding the counts of each hour in a dictionary
dict_labeling_counts = df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).to_dict()
```

```python
# Grouping into Most orders, Average orders and Fewest orders, I decided to use percentile as criteria
list_quantile = list(df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).quantile([0.75, 0.5, 0.25]))
```

```python
#Create loop for changing status of hour
results_labeling_hour = []

for value in df_ords_prods_merged['order_hour_of_day']:
    for key, val in dict_labeling_counts.items():
        if value == key:
            if val < list_quantile[2]:
                results_labeling_hour.append('Fewest orders')
            elif val >= list_quantile[2] and val < list_quantile[1]:
                results_labeling_hour.append('Average orders')
            else:
                results_labeling_hour.append('Most orders')
```

```python
# Create new column and add new defined values
df_ords_prods_merged['Busiest_period_of_day'] = results_labeling_hour
```
